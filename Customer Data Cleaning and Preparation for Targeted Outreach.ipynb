## Data Preprocessing and Cleaning

###  Dropping Colunm "Not_Useful_Column"

```python
df=df.drop(columns="Not_Useful_Column")
df
```

###  Dropping Duplicates Rows

```python
df=df.drop_duplicates()
df
```

### String strip

```python
# Using string strip method for removing 'Special charaters' from column

df["Last_Name"]=df["Last_Name"].str.strip("[./_")
df
```

### Removing Non-alphnumeric characters from 'Phone_Number' column

```python
df["Phone_Number"]=df["Phone_Number"].str.replace(r"[^a-zA-Z0-9]","",regex=True)
df
```

### Normalizing "Phone_Number"

```python
#Formating the phone numbers
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: str(x))
df
```

### Standardizing "Phone_Number"

```python
#Formating the phone numbers
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3] + '-'+ x[3:6] + '-'+ x[6:10] )
df
```

### Handling Missing Values

```python
# Checking for missing values.

df.isnull().sum()
```

```python
# Filling missing values with a specific value.

df.fillna(0)
```

### Removing / replacng Nan,nan,na etc.

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--','')
df
```

### Categorizing "Address"Column into three suc columns

```python
df[["Street_Address","State","Zip_Code"]]=df["Address"].str.split(',', n=2, expand=True)
df
```

### Standerdizing column values all Yes to "Y" and No into "N"

```python
df["Paying Customer"]=df["Paying Customer"].astype(str).str.replace("Yes","Y",regex=True)
df["Paying Customer"]=df["Paying Customer"].astype(str).str.replace("No","N",regex=True)
df["Do_Not_Contact"]=df["Do_Not_Contact"].astype(str).str.replace("Yes","Y",regex=True)
df["Do_Not_Contact"]=df["Do_Not_Contact"].astype(str).str.replace("No","N",regex=True)
df
```

### Removing or replacing N/a and nan 

```python
df=df.replace('N/a','')
df=df.replace('nan','')
df
```

### Fillinf null values with whitespace

```python
df=df.fillna('')
df
```

### Removing unnecessary rows by usning "For loop"

```python
# Dropping row where "Phone_Number" column has no value
for x in df.index:
    if df.loc[x,"Phone_Number"]== '':
        df.drop(x,inplace=True)
df
```

```python
# Dropping row where "Do_Not_Contact" have "Y" value
for x in df.index:
    if df.loc[x,"Do_Not_Contact"]== 'Y':
        df.drop(x,inplace=True)
df    
```

## Final Cleaned and Processed Data

```python
# Resetting the Index No.
df=df.reset_index(drop=True)
df
```
